### 11.0 Regular expression idea

```python
import re
hand = open ('files/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    # search for lines that start with 'From'
    if re.search("^From:",line):
        print(line)
```

### 11.1 Character matching in regular expressions

```python
import re
hand = open ('files/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    # search for lines that start with 'Fxxm:'
    if re.search("^F..m:",line):
        print(line)
```

```python
import re
hand = open ('files/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    # search for lines that start with From and have an at sign
    if re.search("^From:.+@",line):
        print(line)
```

### 11.2 Extracting data using regular expressions

```python
import re
s = "A message from csev@umich.edu to cwen@iupui.edu about meeting @2PM"
lst = re.findall("\S+@\S+",s)
print(lst)
```

```python
import re
hand = open ('files/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    # Extract email address
    x = re.findall("\S+@\S+",line)
    if len(x)>0:
        print(x)
```

```python
import re
hand = open ('files/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    # Search for lines that have an at sign between characters
    # The characters must be a letter or number
    x = re.findall("[a-zA-Z0-9]\S*@\S*[a-zA-Z]",line)
    if len(x)>0:
        print(x)
```

### 11.3 Combining searching and extracting

```python
import re
hand = open ('files/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    # search for lines that start with "X" followed by any non
    # whitespace characters and ":"
    # followed by a space and any number
    # The number can include a decimal.
    if re.search("^X\S*: [0-9.]+",line):
        print(line)
```

```python
import re
hand = open ('files/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    # search for lines that start with "X" followed by any non
    # whitespace characters and ":"
    # followed by a space and any number
    # The number can include a decimal.
    x = re.findall("^X\S*: [0-9.]+",line)
    if len(x)>0:
        print(x)
```

```python
import re
hand = open ('files/mbox-short.txt')
for line in hand:
    line = line.rstrip()
    # search for lines that start with From and a character
    # followed by a two digit number between 00 and 99 followed by ":"
    x= re.findall("^From .* ([0-9][0-9]):",line)
    if len(x)>0:
        print(x)
```

### 11.4 Escape character

```python
import re
x = 'We just received $10.00 for cookies.'
y = re.findall("\$[0-9.]+",x)
print(y)
```

```python
# exe 1
import re
s = input("Enter a regular expression: ")
count = 0
hand = open ('files/mbox.txt')
for line in hand:
    line = line.rstrip()
    # search for lines that start with From and have an at sign
    if re.search(s,line):
        count= count+1
print("mbox.txt had 1798 lines that matched "+s)
```

```python
# exe2
file = input("Enter file: ")
hand = open ('files/'+file)
acc = 0
count = 0
for line in hand:
    line = line.rstrip()
    # search for lines that start with From and a character
    # followed by a two digit number between 00 and 99 followed by ":"
    x= re.findall("New Revision: ([0-9]+)",line)
    if len(x)>0:
        acc = float(x[0])+acc
        count = count +1
print(acc/count)
```
